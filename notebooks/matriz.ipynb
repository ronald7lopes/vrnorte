### Este arquivo serve para mostrar a quantidade de acessos com a matriz para a data selecionada

```python
# Célula de importação das bibliotecas necessárias
import pandas as pd
import seaborn as sns
```

```python
# Criando o Dataframe na variável DF
df = pd.read_csv('..\\data\\matriz_ids.csv')

# Transformando a coluna 'createdDate' em uma coluna de datas
df['createdDate'] = pd.to_datetime(df['createdDate'])

# Criando a coluna 'mês' e 'ano" para uma filtragem mais precisa
df['mes'] = df['createdDate'].dt.month
df['ano'] = df['createdDate'].dt.year
# df['dia'] = df['createdDate'].dt.day     # Caso queira criar uma coluna para os dias


# criando um agrupamento de atendimentos por dia em cada ano

# nome do agrupamento  agrupando por ano e mês    
atendimentos_por_mes = df.groupby(['ano', 'mes']).size().reset_index(name='count')


# tema do Seaborn
sns.set_theme(style="darkgrid")


# Plotagem com Seaborn em linhas
sns.lineplot(x="mes", y="count",
             hue="ano",
             data=atendimentos_por_mes)
```

```python
# Filtrando um novo DataFrame para um ano específico
df_ano = df[df['ano'] == 2024]
```

```python
# Agrupando os atendimentos por analista em cada mês
atendimentos_analista_por_mês = df_ano.groupby(['analista', 'mes']).size().reset_index(name='cont')
```

```python
# Filtre para um mês no ano especificado
atendimentos_analista_por_mês.query('mes == 1')
```

```python
# Plotatem em linhas para comparar o número de atendimentos do analista no ano filtrado

sns.lineplot(x="mes", y="cont",
             hue="analista",
             data=atendimentos_analista_por_mês
)
```
